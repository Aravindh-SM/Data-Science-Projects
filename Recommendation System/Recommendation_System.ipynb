## 4. Build a collaborative filtering model using SVD.

```python
from surprise import Reader,Dataset,SVD,KNNWithMeans,accuracy
from surprise.model_selection import cross_validate,train_test_split
```

```python
# arranging columns in the order of user id,item id and rating to be fed in the svd
columns_titles = ['author','product','score']
vs_rev=r_dup_copy.reindex(columns=columns_titles)
```

```python
# Keeping only 5000 data samples and using random state=612
vs_data = vs_rev.sample(n=5000, random_state=612)
```

```python
reader = Reader(rating_scale=(1, 10))
data = Dataset.load_from_df(vs_data,reader = reader)
```

```python
trainset = data.build_full_trainset()
trainset.ur
```

```python
algo = SVD()
algo.fit(trainset)
```

```python
#Predicting ratings for all pairs (u, i) that are NOT in the training set.
testset = trainset.build_anti_testset()
```

```python
predictions = algo.test(testset)
predictions
```

```python
from collections import defaultdict
#Mapping the predictions to each user
top = defaultdict(list)
for uid, iid, true_r, est, _ in predictions:
    top[uid].append((iid, est))

#Sorting the predictions for each user and retrieve the k highest ones.
for uid, user_ratings in top.items():
    user_ratings.sort(key=lambda x: x[1], reverse=True)
    top[uid] = user_ratings[:5]
    
top
```

```python
#Printing the recommended items for each user
for uid, user_ratings in top.items():
    print(uid, [iid for (iid, _) in user_ratings])
```

```python
#5. Evaluate the collaborative model. Print RMSE value.
print("SVD Model : Test Set")
accuracy.rmse(predictions, verbose=True)
```

```python
#9. Try cross validation techniques to get better results.
cross_validate(algo, data, measures=['RMSE'], cv=3, verbose=False)
#RMSE of SVD model is lower
```

```python
def get_Iu(uid):
    try:
        return len(trainset.ur[trainset.to_inner_uid(uid)])
    except ValueError: # user was not part of the trainset
        return 0
    
def get_Ui(iid):
    try: 
        return len(trainset.ir[trainset.to_inner_iid(iid)])
    except ValueError:
        return 0
    
bf = pd.DataFrame(predictions, columns=['uid', 'iid', 'rui', 'est', 'details'])
bf['Iu'] = bf.uid.apply(get_Iu)
bf['Ui'] = bf.iid.apply(get_Ui)
bf['err'] = abs(bf.est - bf.rui)
best_predictions = bf.sort_values(by='err')[:10]
```

```python
#Build a collaborative filtering model using kNNWithMeans from surprise using Item based model

reader = Reader(rating_scale=(1, 10))
data_I = Dataset.load_from_df(vs_data,reader = reader)
trainset_I, testset_I = train_test_split(data_I, test_size=.15)

# Use user_based true/false to switch between user-based or item-based collaborative filtering

algo = KNNWithMeans(k=50, sim_options={'name': 'pearson_baseline', 'user_based': False})
algo.fit(trainset_I)
```

```python
test_pred_I = algo.test(testset_I)
accuracy.rmse(test_pred_I, verbose=True)
```

```python
#Build a collaborative filtering model using kNNWithMeans from surprise using User based model

reader = Reader(rating_scale=(1, 10))
data_U = Dataset.load_from_df(vs_data,reader = reader)
trainset_U, testset_U = train_test_split(data_U, test_size=.15)
algo = KNNWithMeans(k=50, sim_options={'name': 'pearson_baseline', 'user_based': True})
algo.fit(trainset_U)
```

10. Popularity based Recommendation Systems should be used when user interaction of the particular individual data is unavailable. This is also usually done in cold start scenario.
In that case, recommending something based on just popularity is better than no recommendation.

11. Collaborative filtering based Recommendation is done to build recommenders that give suggestions to a user on the basis of the likes and dislikes of similar users.  
This can be done when there is adequate user interaction info about the particular individual and many other users, which is sufficient for correlating the interactions and clustering the users.

12. Content based and hybrid recommendation systems could also be used for this scenario.
